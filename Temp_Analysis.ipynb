```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_temp = {
    'Days' : [1, 2, 3, 4, 5, 6, 7],
    'Temp_Sudan' : [22.5, 22.7, 25, 27.3, 26.1, 22.4, 20,],
    'Temp_Nyala' : [30.2, 29, 25, 33.3, 39.7, 35, 28],
    }
```

```python
df_3 = pd.DataFrame(data_temp)
print(df_3)
```

```python
# Temp Of Sudan
```

```python
print(df_3[['Days','Temp_Sudan']])
```

```python
print(df_3[['Days','Temp_Nyala']])
```

```python
#Line Plots
```

```python
plt.plot(df_3['Days'],df_3['Temp_Sudan'])
plt.show()
```

```python
plt.plot(df_3['Days'],df_3['Temp_Nyala'])
plt.show()
```

```python
# Scatter Plots
```

```python
plt.scatter(df_3['Days'],df_3['Temp_Sudan'])
plt.show()
```

```python
# Bar Charts
```

```python
plt.bar(df_3['Days'],df_3['Temp_Nyala'])
plt.show()
```

```python
# Histograms
```

```python
plt.hist(data_temp, bins = 2)
plt.show()
```

```python
# x-axis & y-axis

plt.plot(df_3['Days'],df_3['Temp_Sudan'])
plt.xlabel('Days of week')
plt.ylabel('Temperatures in Cell.')

plt.title('Temperatures Degrees in Sudan')
plt.show()
```

```python
plt.plot(df_3['Days'],df_3['Temp_Nyala'])
plt.xlabel('Days of week')
plt.ylabel('Temperatures in Cell.')

plt.title('Temperatures Degrees in Nyala')
plt.show()
```

```python
# Customization Plots
```

```python
# x-axis & y-axis

plt.plot(df_3['Days'],df_3['Temp_Sudan'], marker = '*')
plt.plot(df_3['Days'],df_3['Temp_Nyala'], c = 'r', alpha = 0.4, marker = 'o', ls = '--')

# plt.plot(df_3['Days'],df_3['Temp_Sudan'], lablel = 'Sudan') >> You can add lablel in order to know which color can discrable temp

plt.xlabel('Days of week')
plt.ylabel('Temperatures in Cell.')

plt.title('Temperatures Degrees in Sudan, Nyala')
plt.legend(['Sudan','Nyala'])

plt.show()
```

```python
# Subplots
```

```python
figure, axes = plt.subplots(1, 2, figsize = (10, 5))

axes[0].plot(df_3['Days'],df_3['Temp_Sudan'], c = 'r')
axes[0].set_title('Temperatures Degrees in Sudan')

axes[1].plot(df_3['Days'],df_3['Temp_Nyala'], c = 'k')
axes[1].set_title('Temperatures Degrees in Nyala')

plt.savefig('temp.png')

plt.show()
```
